```python
x = [4,6,8,10,12]
fair = [0.45, 0.47, 0.48, 0.47, 0.46]
fig, ax = plt.subplots()
ax.plot(x,fair)
plt.xlabel('Slowdown Speed (m/s)', fontsize = 16)
plt.ylabel('Fairness Factor', fontsize = 16)
loc = plticker.MultipleLocator(base=2.0) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
plt.show()
```
